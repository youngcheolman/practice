## 결측치 전처리


waist와 heap은 bmi로 평균적인 추정이 가능하기 때문에 2개의 항목은 삭제 

loss_part 원핫 인코딩처럼 1~9까지 항목을 더 만들어 분류

diet_purpose는 대부분의 사람들이 원하는 부위가 비슷하다는 판단에 평균으로 대체

diet와 dise는 사람들의 개인적인 경험과 질병을 표시한 데이터기 때문에 결측값들은 그냥 0으로 표기 하여 나타내려고 함.

```python
df.drop(['waist','heap'], axis=1, inplace=True)
```

```python
#diet_purpose의 결측값을 최빈값 1로 대체
df["diet_purpose"].fillna("1", inplace = True)
df["dise"].fillna("0", inplace = True)
df["loss_part"].fillna("0", inplace = True)
```

```python
df
```

```python
#문자열 데이터를 수치화 할 수 없기에 ,를 삭제

df['diet'] = df['diet'].str.replace(",","")
df
```

```python
# diet 3,4,5,6 이 포함되어 있으면 1 없으면 0으로 표기

df['diet'] = df['diet'].map(lambda x : str(x))
diet_experience = [0 for i in df['diet']]
for i, diet in enumerate(df['diet']):
  if ('3' in diet) | ('4' in diet) | ('5' in diet)| ('6' in diet):
      #print(diet)
      diet_experience[i] = 1
df['diet_experience'] = diet_experience
```

```python
#loss_part데이터를 1~9까지 새로운 columns을 하나씩 만듬

df["loss_part_1"] = df["loss_part"].str.contains("1", na = False)*1
df["loss_part_2"] = df["loss_part"].str.contains("2", na = False)*1
df["loss_part_3"] = df["loss_part"].str.contains("3", na = False)*1
df["loss_part_4"] = df["loss_part"].str.contains("4", na = False)*1
df["loss_part_5"] = df["loss_part"].str.contains("5", na = False)*1
df["loss_part_6"] = df["loss_part"].str.contains("6", na = False)*1
df["loss_part_7"] = df["loss_part"].str.contains("7", na = False)*1
df["loss_part_8"] = df["loss_part"].str.contains("8", na = False)*1
df["loss_part_9"] = df["loss_part"].str.contains("9", na = False)*1
```

```python
df
```

```python
df["height"].fillna(df["height"].mean(),inplace = True)
df["weight"].fillna(df["weight"].mean(),inplace = True)
df["bmi"].fillna(df["bmi"].mean(),inplace = True)
```

```python
df.dropna(axis=0, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report

X = df[['occupation','diet_purpose', 'bmi', 'p_price', 'p_quantity','product buy season',
       'diet_experience',"join season",'loss_part_1', 'loss_part_2', 'loss_part_3', 'loss_part_4',
       'loss_part_5', 'loss_part_6', 'loss_part_7', 'loss_part_8','loss_part_9']]
Y = df['label']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

smote = SMOTE(random_state = 0)
X_train, Y_train = smote.fit_resample(X_train, Y_train)

from sklearn.tree import DecisionTreeClassifier

DT_model = DecisionTreeClassifier(random_state = 0, max_depth = 3)
DT_model.fit(X_train, Y_train)
Y_pred = DT_model.predict(X_test)
print("accuracy : {:.3f}".format(DT_model.score(X_train, Y_train)))
print(classification_report(Y_test, Y_pred))
```

```python
feature_name = X.columns
feature_importance = DT_model.feature_importances_
sorted(zip(feature_importance,feature_name), reverse = True)
```

#랜덤 포레스트

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

#1.변수선택
X = df[['occupation','diet_purpose', 'bmi', 'p_price', 'p_quantity','product buy season',
       'diet_experience',"join season",'loss_part_1', 'loss_part_2', 'loss_part_3', 'loss_part_4',
       'loss_part_5', 'loss_part_6', 'loss_part_7', 'loss_part_8','loss_part_9']]
Y = df['label']
#2.train - test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

#3.데이터 전처리(preprocessing)
ct = ColumnTransformer([("scaling", StandardScaler(), ['bmi','p_price', 'p_quantity']),
                        ("onehot", OneHotEncoder(sparse = False), ['occupation','diet_purpose','diet_experience',"join season",'product buy season','loss_part_1', 'loss_part_2', 'loss_part_3', 'loss_part_4',
       'loss_part_5', 'loss_part_6', 'loss_part_7', 'loss_part_8','loss_part_9'])])
ct.fit(X)
X_train = ct.transform(X_train)
X_test = ct.transform(X_test)

#4. 오버샘플링
smote = SMOTE(random_state=0)
X_train, Y_train = smote.fit_resample(X_train, Y_train)
```

```python
#5. 모형 생성
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state = 0, n_estimators= 300, max_depth = 3)
model. fit(X_train,Y_train)
Y_pred = model.predict(X_test)
print("Y_predicti value: \n", Y_pred)
print("accuracy(test) : {:.3f}".format(model.score(X_test, Y_test)))
print(classification_report(Y_test, Y_pred))
```

```python
feature_name = X.columns
feature_importance = model.feature_importances_
sorted(zip(feature_importance,feature_name), reverse = True)
```

#KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder


X = df[['occupation', 'diet_purpose', 'bmi', 'p_price', 'p_quantity','product buy season',
       'diet_experience',"join season",'loss_part_1', 'loss_part_2', 'loss_part_3', 'loss_part_4',
       'loss_part_5', 'loss_part_6', 'loss_part_7', 'loss_part_8',
       'loss_part_9']]
Y = df['label']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)

scaler = StandardScaler().fit(X_train, Y_train)
X_test = scaler.transform(X_test)
X_train = scaler.transform(X_train)

X_train, Y_train = SMOTE(random_state = 0).fit_resample(X_train, Y_train)

from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors = 9, metric = "euclidean") #euclidean - 두점사이 거리 공식

knn_model.fit(X_train, Y_train)
Y_pred = knn_model.predict(X_test)
print("accuracy : {:.3f}".format(knn_model.score(X_train, Y_train)))
print(classification_report(Y_test, Y_pred)) #accuracy, f1 score 보는게 가장 정확
#실제 f1 score : 아래의 0과 1 f1 score의 평균값
#맨위 accuracy - train 값 (train 데이터 잘 설명 하는지)
#아래 accuracy - test pred 계산했을때 나온 값 (예측잘 하는지)
```
