```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/감귤2.csv", encoding= "cp949") #감귤 파일 불러오기
```

```python
df = df.dropna(axis=1) # none 값 삭제
```

```python
df.columns = ['날짜',	'품목',	'등급',	'거래단위',	'최저가',	'최고가',	'평균가',	'전일대비등락', '전일평균대비'] #데이터 칼럼을 보기 좋게 다시 정리
df = df.iloc[0:53,0:8] #불필요하다고 판단한 '전일평균대비' 항목 삭제
```
